```python
# 4. é…ç½® config.tomlï¼ˆè¿™é‡Œç”¨äº¤äº’å¼è¾“å…¥ API Keyï¼Œæ›´å®‰å…¨ï¼‰
# ä½ éœ€è¦å‡†å¤‡ï¼š
# - Pexels API Key: https://www.pexels.com/api/ å…è´¹æ³¨å†Œè·å–
# - LLM API Key: æ¨è DeepSeekï¼ˆå…è´¹é¢åº¦é«˜ï¼‰ï¼šhttps://platform.deepseek.com/api-keys è·å–

import os
from toml import load, dump

config_path = "config.toml"

# å¦‚æœå·²æœ‰ config.tomlï¼ŒåŠ è½½ï¼›å¦åˆ™ä»ç¤ºä¾‹å¤åˆ¶
if not os.path.exists(config_path):
    !cp config.example.toml config.toml

with open(config_path, "r", encoding="utf-8") as f:
    config = load(f)

print("è¯·ä¾æ¬¡è¾“å…¥ä»¥ä¸‹ä¿¡æ¯ï¼ˆå›è½¦ç¡®è®¤ï¼‰ï¼š")

# Pexels Key
pexels_key = input("Pexels API Key: ").strip()
if pexels_key:
    config["pexels_api_keys"] = [pexels_key]

# LLM é€‰æ‹©ï¼ˆæ¨è deepseekï¼Œå›½å†…å¿«ï¼‰
llm_provider = input("LLM æä¾›å•† (æ¨è deepseek / moonshot / openai): ").strip().lower() or "deepseek"
config["llm_provider"] = llm_provider

if llm_provider == "deepseek":
    deepseek_key = input("DeepSeek API Key: ").strip()
    config["deepseek_api_key"] = deepseek_key
elif llm_provider == "openai":
    openai_key = input("OpenAI API Key: ").strip()
    config["openai_api_key"] = openai_key
# å¯ä»¥ç»§ç»­æ·»åŠ å…¶ä»–æ¨¡å‹æ”¯æŒ...

# ä¿å­˜é…ç½®
with open(config_path, "w", encoding="utf-8") as f:
    dump(config, f)

print("config.toml é…ç½®å®Œæˆï¼")
```
